```python
import pandas as pd 
import numpy as np
```

```python
data = {
    '日期': pd.date_range('2024-01-01', periods=10, freq='D'),
    '销售额': [120, 150, 90, 200, 170, 300, 50, 400, 210, 180],
    '订单数': [12, 15, 9, 20, 17, 30, 5, 40, 21, 18],
    '客服组': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}
```

```python
df=pd.DataFrame(data)
```

```python
print (df['销售额'].describe())
print("标准差：",df['销售额'].std())
```

```python
print(df.groupby('客服组')['销售额'].sum())
```

```python
mean = df['销售额'].mean()
std = df['销售额'].std()
df['异常标记']=np.where(df['销售额']<mean-2*std,'异常低',
                    np.where(df['销售额'] > mean+2*std,'异常高','正常'))
print(df[['日期','销售额','异常标记']])
```
